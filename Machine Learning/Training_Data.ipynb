```python
import pandas as pd
import random
```

```python
data = []
```

```python
def generateData(tempL, tempH, hrL, hrH, spo2L, spo2H, levelL, levelH):
    for i in range(1000):
        row=[]
        temp=random.uniform(tempL, tempH)
        hr=random.randint(hrL, hrH)
        spo2=random.uniform(spo2L, spo2H)
        level=random.randint(levelL, levelH)
        row.append(temp)
        row.append(hr)    
        row.append(spo2)    
        row.append(level)
        data.append(row)
```

```python
generateData(98.0, 99.0, 70, 90, 98.5, 100.0, 90.0, 100.0)
generateData(96.5, 99.5, 60, 70, 95.0, 100.0, 80.0, 90.0)
generateData(96.5, 99.5, 90, 100, 95.0, 100.0, 80.0, 90.0)
generateData(97.0, 99.0, 60, 100, 95.0, 100.0, 90.0, 100.0)
generateData(95.0, 97.0, 60, 100, 97.0, 100.0, 80.0, 90.0)
generateData(97.0, 100.0, 60, 100, 97.0, 100.0, 75.0, 85.0)
generateData(99.0, 100.5, 100, 120, 90.0, 94.0, 75.0, 85.0)
generateData(100.5, 102.5, 100, 110, 90.0, 94.0, 70.0, 80.0)
generateData(94.0, 95.0, 100, 120, 90.0, 94.0, 70.0, 80.0)
generateData(94.0, 95.0, 100, 120, 85.0, 90.0, 70.0, 80.0)
```

```python
dataDF = pd.DataFrame(data)
dataDF.head()
```
